```python
import yfinance as yf
```

```python
def get_data(ticker_list):
    data = yf.download(tickers=' '.join(ticker_list),
                       period='ytd',
                       group_by='ticker',
                       auto_adjust = True,
                       prepost = True,
                       threads = True,)
    return(data)
lst = ['SPY', 'AAPL']
data = get_data(lst)
```
